```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By # use for locating elements
from selenium.webdriver.support import expected_conditions as EC
import time

#specify the path to your WebDriver executable
#replace 'path/to/your/chromedriver' with the actual path
DRIVER_PATH = './chromedriver-win64/chromedriver.exe'  # assuming chromedriver is in the same directory

#set up the services
#this is the modern way to initialize the driver
service = Service(executable_path=DRIVER_PATH)
# initialize the WebDriver
driver = webdriver.Chrome(service=service)
wait = WebDriverWait(driver, 20)
# start automation

#navigate to a webpage
print('Navigating to website...')
driver.get('https://www.duckduckgo.com')

## get and print the page title
page_title = driver.title
print(f'Page Title> {page_title}')

# find the search box and type a query
#locate an element by its 'name' attribute which is 'q' on Google
#search_box = driver.find_element(By.NAME, 'q')
#search_box.send_keys('What is Selenium?')

graph_image = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/main/article/div[1]/div[1]/div[2]/div/header/div/section[1]/a/img')))
graph_image.click()
##submit the form - the search box is inside a form
#search_box.submit()
graph_image = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div[5]/section/div/div[2]/div/div/button[1]' )))
graph_image.click()

graph_image = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div[5]/section/div/div[2]/div/div/button[2]')))
graph_image.click()

#pause for a moment to see the results (optional, but good for testing)
time.sleep(5)

###---- end automation ---- ###
#close the browser

print('Closing the browser...')
driver.quit()
```
